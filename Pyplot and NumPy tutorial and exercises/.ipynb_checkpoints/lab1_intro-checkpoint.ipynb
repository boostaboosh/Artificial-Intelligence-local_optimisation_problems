## Excercise 2: Functions

The Euclidean distance is the "ordinary" straight-line distance between two points in Euclidean space. Given two points in Cartesian coordinates in the plane, $(x_1, y_1)$ and $(x_2, y_2)$, the Euclidean distance $d$ between them is given by the formula:

$d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}$

Your task is to implement a function called distance, that receives four paramenters the coordinates of two points in the order: $x_1, y_1, x_2, y_2$, and calculates the Euclidean distance between the points $(x_1, y_1)$  and $(x_2, y_2)$

<b>Hint</b>: The math function for square root is 'sqrt'.

```python
# Insert your code for exercise 2 here
import math

def distance(x1, y1, x2, y2):
    dx = (x2 - x1)**2
    dy = (y2 - y1)**2
    d = math.sqrt(dx + dy)
    return d

print(distance(1, 1, 3, 5))
```

```python
# Insert your code  here
import random as rnd
import matplotlib.pyplot as plt
# This line allows to show the plots without having to write plt.show() everytime
%matplotlib inline  

def randomIntList(length):
    list = []
    for counter in range(length):
        number = rnd.randint(0, 100)
        list.append(number)
    return list

list1 = randomIntList(30)
list2 = randomIntList(30)

plt.title("scatter plot")
plt.ylabel("metric 1")
plt.xlabel("metric 2")
plt.scatter(list1, list2, color = "lightblue", linewidths = 1, label = "random")
plt.legend()
plt.show()
#plt.figure()

plt.title("box plot")
plt.ylabel("metric uno")
plt.xlabel("metric due")
boxplotFigure = plt.boxplot([list1, list2])
plt.figure()
```

### Programming Task 2

Implement a function called 'surprise_pizza' that receives as an input an integer (from 0 to 6) indicating the number of desired toppings 'nt'. The function shouldand returns a randomly generated binary digit list with a number 'nt' of 1s. That is, a random selection of 'nt' toppings.

```python
import random as rnd
import numpy as np

def surprise_pizza(nt):
    surpriseToppings = np.zeros(6, dtype = int)
    numToppings = 0
    while numToppings < nt:
        randomPosition = rnd.randint(0, 5)
        if surpriseToppings[randomPosition] != 1:
            surpriseToppings[randomPosition] = 1
            numToppings = numToppings + 1
    return surpriseToppings
```

<b>Question 2</b>: Calculate and report (using the two functions implemented) the prices of 4 surprise pizzas with 2 toppings. 

Print both the binary list with choices and the price of the pizza

```python
# Add your answer here
surprisePizza1 = surprise_pizza(2)
surprisePizza2 = surprise_pizza(2)
pricePizza1 = pizza_price(surprisePizza1)
pricePizza2 = pizza_price(surprisePizza2)
print(f"price pizza 1 {surprisePizza1}: {pricePizza1}")
print(f"price pizza 2 {surprisePizza2}: {pricePizza2}")
```
